## Q2.B
## Weighted Restaurant Rating

```python
#reading file and filtering the data
data=pd.read_csv('C:/Users/ANIKET/Desktop/ANIKET/CN/Zomato.csv',encoding='latin-1')
data=data[data['Country Code']==1]

#creating variable lists
#wrr is weighted restaurant rating
locality=[]
wrr=[]

#calculating the unique localities and then iterating over each locality one by one.
for i in data.Locality.unique():
    
    #creating a copy of dataframe
    df=data.copy()

    locality.append(i)
    #filtering data for a particular locality
    df=df[df.Locality==i]
    
    #multiplying number of votes with the aggregate rating for all restaurants in that locality 
    df['Aggregate rating']=(df.Votes)*(df['Aggregate rating'])
    
    #taking care of case where votes=0
    if (df.sum(axis=0)['Votes'])==0:
        wrr.append(0)
    else:
        #calculating wrr
        wr=(df.sum(axis=0)['Aggregate rating'])/(df.sum(axis=0)['Votes'])
        wrr.append(wr)

#code for plotting the required bar graph
import numpy as np
wrr=np.array(wrr)
locality=np.array(locality)
index=np.argsort(wrr)
locality=locality[index]
wrr=wrr[index]

l=[]
r=[]
#finding top 10 
for i in range(-1,-11,-1):
    l.append(locality[i])
    r.append(wrr[i])
    
plt.bar(l,r)
plt.xticks(rotation=90)
plt.xlabel('NAME OF LOCALITY')
plt.ylabel('WEIGHTED RESTAURANT RATING')
plt.grid()
plt.show()
```

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#reading the file
data=pd.read_csv('C:/Users/ANIKET/Desktop/ANIKET/CN/Zomato.csv',encoding='latin-1')

#filtering the data for country India
data=data[data['Country Code']==1]

#a contains the restaurants in the data
a=data['Restaurant Name']

#b contains the cuisines in the data
b=data['Cuisines']

#c contains the aggregate rating in the data
c=data['Aggregate rating']

#r variable denotes those restaurants that serve only 1 cuisine
r=[]

#s variable denotes those restaurants that serve multiple cuisines
s=[]

#x variable containes the rating of corresponding restaurants in variable r
x=[]

#y variable containes the rating of corresponding restaurants in variable s
y=[]

#we have plotted the graph for first 50 Indian restaurants in the data.
for i in range(50):
    
    #splitting the cuisines so as to calculate the number of cuisines i.e. the length. 
    j=b[b.index[i]].split(',')
    
    if len(j)==1:
        r.append(a[a.index[i]])
        x.append(c[c.index[i]])
        
    else:
        s.append(a[a.index[i]])
        y.append(c[c.index[i]])

#Plotting graph
plt.plot(r,x,label='Restaurants serving only one specific cuisine')
plt.plot(s,y,label='Restaurants serving more than one cuisine')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.plot()

#INFERENCE
#As can be seen from the plot,
#that the restaurants which serve only a specific cusine have lower rating compared to those who serve variety of cuisines.
```

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#reading the file
data=pd.read_csv('C:/Users/ANIKET/Desktop/ANIKET/CN/Zomato.csv',encoding='latin-1')

#sorting data for country india
data=data[data['Country Code']==1]
vote=[]
r_name=[]
z=len(data['Restaurant Name'].value_counts())

#iterating over the restaurants
for i in range(z):
    #making a copy of data
    df=data.copy()
    
    #filtering the data for a restaurant
    name=df['Restaurant Name'].value_counts().index[i]
    
    r_name.append(name)
    df=df[df['Restaurant Name']==name]
    vote.append(df.sum(axis=0,skipna=True)['Votes'])
    
#sorting using argsort
index=np.argsort(vote)
vote=np.array(vote)
r_name=np.array(r_name)
vote=vote[index]
r_name=r_name[index]
v=[]
r=[]

#collecting top 10
for i in range(-1,-11,-1):
    v.append(vote[i])
    r.append(r_name[i])
    
#plotting the graph
plt.bar(r,v)
plt.xticks(rotation=90)
plt.xlabel('RESTAURANT')
plt.ylabel('VOTES')
plt.show()
```

```python
#importing libraries and reading the dataset and convert into dataframe
import pandas as pd
data=pd.read_csv('C:/Users/ANIKET/Desktop/ANIKET/CN/Zomato.csv',encoding='latin-1')

#filtering the data for country India
data=data[data['Country Code']==1]
city=[]
wrr=[]
count=[]

#calculating weighted restaurant rating (wrr) for all the restaurants present in a city
for j in range(43):
    df=data.copy()
    
    #selecting a particular city
    i=df.City.value_counts().index[j]
    city.append(i)
    
    #filter data for particular city
    df=df[df.City==i]
    count.append(df.City.value_counts()[i])
    
    #calculate weighted rating
    df['Aggregate rating']=(df.Votes)*(df['Aggregate rating'])
    wr=(df.sum(axis=0)['Aggregate rating'])/(df.sum(axis=0)['Votes'])
    wrr.append(wr)

#plot bubble chart
plt.scatter(city,count,s=wrr)
plt.xticks(rotation=90)
plt.title('Bubble chart')
plt.show()
```
